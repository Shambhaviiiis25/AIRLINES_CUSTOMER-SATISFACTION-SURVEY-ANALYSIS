# Pre-Flight Analysis

**Q1: Determine the Overall percentage of positive, negative or neutral ratings.**

```python
#Filter satisfied customers
satdf1=df1[df1["satisfaction"]=="satisfied"]
# The count of satisfied customers
sat_pre=satdf1.value_counts(dropna=False).sum()
# Filter dissatisfied or neutral customers
disdf1=df1[df1["satisfaction"]=="neutral or dissatisfied"]
# The count of dissatisfied customers
dis_pre=disdf1.value_counts(dropna=False).sum()
#Total number of customers
total_customer=df1.value_counts(dropna=False).sum()
# % of satisfied
per_sat=(sat_pre/total_customer)*100
per_dis=(dis_pre/total_customer)*100
print("The satisfied customers in Pre-flight factors is:",per_sat,
      "The dissatisfied customers in Pre-flight factors is:",per_dis)
```

**Q7: To check average satisfaction by Age group**

```python
# Group by Age segments and calculate mean satisfaction for in-flight factors
age_segments = ['Children', 'Young Adult', 'Adult', 'Senior']

# Define the in-flight factors of interest
in_flight_factors = ['Inflight entertainment', 'Seat comfort', 'Food and drink', 'Cleanliness','On-board service']

# Calculate mean satisfaction per age group
df['Age Group'] = pd.cut(df['Age'], bins=[7, 17, 24, 55, np.inf], labels=age_segments)
age_group_means = df.groupby('Age Group')[in_flight_factors].mean()

# Plot the results
age_group_means.plot(kind='bar', figsize=(10, 6))
plt.title("Average Satisfaction for In-flight Factors by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Satisfaction Rating")
plt.xticks(rotation=0)
plt.legend(title="In-flight Factor", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

**Q8:To check customer satisfaction based on class: "Business," "Eco Plus," and "Eco"**

```python
# Filter by Type of Travel
business = df[df["Class"] == "Business"]
eco_plus = df[df["Class"] == "Eco Plus"]
eco = df[df["Class"] == "Eco"]

# Further segment by satisfaction for each class
bus_sat = business[business["satisfaction"] == "satisfied"]
bus_dis = business[business["satisfaction"] == "neutral or dissatisfied"]

eco_plus_sat = eco_plus[eco_plus["satisfaction"] == "satisfied"]
eco_plus_dis = eco_plus[eco_plus["satisfaction"] == "neutral or dissatisfied"]

eco_sat = eco[eco["satisfaction"] == "satisfied"]
eco_dis = eco[eco["satisfaction"] == "neutral or dissatisfied"]

# Total count for each class
bus_total = business.value_counts().sum()
eco_plus_total = eco_plus.value_counts().sum()
eco_total = eco.value_counts().sum()

# Calculate percentages
bus_sat_per = (bus_sat.value_counts().sum() / bus_total) * 100
bus_dis_per = (bus_dis.value_counts().sum() / bus_total) * 100

eco_plus_sat_per = (eco_plus_sat.value_counts().sum() / eco_plus_total) * 100
eco_plus_dis_per = (eco_plus_dis.value_counts().sum() / eco_plus_total) * 100

eco_sat_per = (eco_sat.value_counts().sum() / eco_total) * 100
eco_dis_per = (eco_dis.value_counts().sum() / eco_total) * 100

# Create a DataFrame with the results
data = {
    'Class': ['Business', 'Eco Plus', 'Eco'],
    'Satisfied Count': [bus_sat.value_counts().sum(), eco_plus_sat.value_counts().sum(), eco_sat.value_counts().sum()],
    'Dissatisfied Count': [bus_dis.value_counts().sum(), eco_plus_dis.value_counts().sum(), eco_dis.value_counts().sum()],
    'Total Count': [bus_total, eco_plus_total, eco_total],
    '% Satisfied': [bus_sat_per, eco_plus_sat_per, eco_sat_per],
    '% Dissatisfied': [bus_dis_per, eco_plus_dis_per, eco_dis_per]
}

df_segmented = pd.DataFrame(data)

# Display the DataFrame
df_segmented
```

**Q9: To check customer satisfaction with "Baggage handling"**

```python
def postflight(Rating):
    if Rating >= 4:
        return "Good"
    elif Rating == 3:
        return "Average"
    elif Rating <= 2:
        return "Bad"

# Apply the 'postflight' function to create the 'Rating' column in df3
df3["Rating"] = df3["Baggage handling"].apply(postflight)

# Calculate counts for neutral or dissatisfied loyal customers
good = df3[df3["Rating"]=="Good"].value_counts().sum()
average = df3[df3["Rating"]=="Average"].value_counts().sum()
bad= df3[df3["Rating"]=="Bad"].value_counts().sum()
#Percentage
good_per=(good/totalcustomer)*100
average_per=(average/totalcustomer)*100
bad_per=(bad/totalcustomer)*100
print("Good:",good_per)
print("Average:",average_per)
print("Bad:",bad_per)
```

**Q10: To check class type with customers who have rated "Baggage Handling" less than 3**

```python
# Filtering
badrating = df3[df3["Rating"] == "Bad"]

# Business Class
bad_business = badrating[badrating["Class"] == "Business"].shape[0]
# Eco Plus Class
bad_ecoplus = badrating[badrating["Class"] == "Eco Plus"].shape[0]
# Eco Class
bad_eco = badrating[badrating["Class"] == "Eco"].shape[0]

# Total number of "Bad" ratings across all classes
total_bad = bad_business + bad_ecoplus + bad_eco

# Calculating percentage for each class
bad_business_percent = (bad_business / total_bad) * 100 
bad_ecoplus_percent = (bad_ecoplus / total_bad) * 100 
bad_eco_percent = (bad_eco / total_bad) * 100

# Creating a dataframe
datadf3 = {
    'Class': ['Business', 'Eco Plus', 'Eco'],
    'Total Bad Count': [bad_business, bad_ecoplus, bad_eco],
    '% of each class': [bad_business_percent, bad_ecoplus_percent, bad_eco_percent]
}
df3_class_bad = pd.DataFrame(datadf3)
df3_class_bad = df3_class_bad.sort_values(by='Total Bad Count', ascending=False)
df3_class_bad
```

**Q11: To check Male Vs Female customers who have rated "Baggage Handling" less than 3**

```python
# Male
male = badrating[badrating["Gender"] == "Male"].shape[0]
# Female
female = badrating[badrating["Gender"] == "Female"].shape[0]

# Calculating percentage for both male and female
male_percent = (male / total_bad) * 100 
female_percent = (female/ total_bad) * 100 

# Creating a dataframe
datadf3_gender = {
    'Class': ['Male','Female'],
    'Total Bad Count': [male,female],
    '% ': [male_percent, female_percent]
}
df3_bad = pd.DataFrame(datadf3_gender)
df3_bad

```
